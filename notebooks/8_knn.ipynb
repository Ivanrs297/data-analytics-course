# Intro to KNN

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=100, n_features=2, centers=2, random_state=3, cluster_std = 5)
X.shape, y.shape
```

```python
type(X), type(y)
```

```python
y[y == 1]
```

```python
y[y == 0]
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:,0], X[:,1], c = y == 0, label = "Class 0")
plt.scatter(X[:,0], X[:,1], c = y == 1, label = "Class 1")
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Data")
plt.legend()
```

# Model

```python
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier()
```

```python
clf.fit(X, y)
```

```python
clf.score(X, y)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
X_train.shape, X_test.shape
```

```python
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```
